# Contour Plots
***

To make a contour plot, we first need a set of data on a 2D grid. 

<img src="../images/plotting-contour.png" style="float:left; padding-right: 30px" width=550>

For example, think of height plots. For every (x, y) coordinate on the map, we are given a value of the height.

Therefore, we need a list of (x, y) coordinates and their heights. The easiest way to represent this data is on a grid, like this one:

<img src="../images/plotting-contour-excel.png" width=350>

> **Notice**: the y-values increase downwards. So when you make the contour plot, the **bottom row will be on the top**.

For best results, make sure the **grid is filled out completely**. If you are missing some cell values, the contour plot will skip these places rather than trying to draw a contour over it.


## Functions we'll be using
***

#### 1. `np.genfromtxt` to load in the file 

Instead of using **pandas** to import the csv, we will use **NumPy** directly, since we have a grid of numbers rather than a meaningful data table. <br>
NumPy has a function `genfromtxt(path, delimiter, ...)` that creates a NumPy array from a text file, similar to pandas' load_csv.

#### 2. `plt.contour()` or `plt.contourf()` to make a contour plot

`contour()` creates a normal contour plot, and `contourf()` creates a **filled** contour plot with colors instead of lines. It's up to you which one to use. 

Both are used very simply: `contour(data, add. args)`. You will likely not need to use any additional arguments, but some of them are:

* `levels=`: sets specific contour levels. If `levels` is not given, matplotlib decides on the optimal levels itself.
* `vmin=`, `vmax=`: the min/max of the contour levels. E.g. if your mountain is not taller than 6000m, you can set `vmax=6000` to ignore any contours higher than that.
* `linewidths=` width of the contour lines. If set with 1 number, applies to all lines; but you can edit individual contours too.
* `cmap=` color map used. The default one is **viridis** (and quite pretty, too), but you can play with [some other available colormaps](https://matplotlib.org/examples/color/colormaps_reference.html)

#### 3. `plt.colorbar()` to display the color bar

In addition, it is often useful to display the **colorbar** on the plot. It's a third axis that shows you which value does each color correspond to. You can show a colorbar using `plt.colorbar()`

***

First, import stuff:

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

Now, make a simple contour plot, without setting anything fancy. We use the same data as on the Excel screenshot above.

```python
# It's important to set the delimiter here; default for numpy is not ","
data = np.genfromtxt("../data/plotting-contour.csv", delimiter=",")   

plt.contourf(data)  # Create a filled contour plot
plt.colorbar()      # Show a colorbar
```

Here is what happens if the CSV file is missing some data:

```python
# It's important to set the delimiter here; default for numpy is not ","
data = np.genfromtxt("../data/plotting-contour-missing.csv", delimiter=",")   

plt.contourf(data)  # Create a filled contour plot
plt.colorbar()      # Show a colorbar
```

And finally, here is a plot with custom contour lines and a different color map.

```python
# It's important to set the delimiter here; default for numpy is not ","
data = np.genfromtxt("../data/plotting-contour.csv", delimiter=",")   

levels = np.linspace(0, 7, 20)                            # Create 20 contour values from 0 to 7
plt.contour(data, levels=levels, cmap="inferno")          # Create an ufilled contour plot with an inferno colormap
plt.colorbar()                                            # Show a colorbar
```
