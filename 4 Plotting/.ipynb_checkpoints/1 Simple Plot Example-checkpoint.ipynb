# Simple `plot()` Command
***

The simplest plotting function in Python is `plot()` and it is used in most other introductory course. If you just want to plot a bunch of x-values against a bunch of y-values and maybe connect them with lines, `plot()` is the way to go!

First, we import the **pyplot** module of **matplotlib**.

```python
import matplotlib.pyplot as plt   # Import matplotlib
# Tell Jupyter to display plots inside the notebook
%matplotlib inline                
```

[Here is some data](https://trends.collegeboard.org/college-pricing/figures-tables/tuition-fees-room-board-over-time) from a study of how tuition changes with time.

```python
years   = [1988,  1998,  2008,  2018]
tuition = [17000, 22710, 28440, 35830] 
```

We can use the `plot()` command to plot it. `plot()` has 2 required parameters:

    plot(xdata, ydata)
    
* `xdata` is the x-values you want to plot
* `ydata` is the y-values

This will make a very simple plot with all the default settings.

```python
plt.plot(years, tuition)
```

You can customize the plot with additional parameters in the `plot()` function. If you do, the order of parameters is:

    plot(xdata, ydata, format, any other parameters)
    
* `format` is a format string that tells MatPlotLib how to display the data. <br>
   [See the plot formatting tutorial on how to use it](3%20Formatting%20Guide.ipynb)

There are lots of additional parameters; you can [see a full description in MPL docs](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.plot.html#matplotlib.pyplot.plot). Some common ones are:

* `lw` or `linewidth` sets the **line width** to some number. By default, `lw=1`.
* `ms` or `markersize` sets thr **marker size** of the markers, if you have them. By fault, `ms=6`.
* `color` sets the **color** of the line and the markers. See different ways to set the color in the [formatting guide](3%20Formatting%20Guide.ipynb)
* `label` gives the data set a name, in case you want to create a legend. See the [legends and labels guide](4%20Legends%2C%20Labels%2C%20Annotations.ipynb)

```python
plt.plot(years, tuition, "ko-", lw=3, ms=16, color='#aa0000')
```

## Issues with using `plot()`

`plot()` is great for some quick and dity plots, but it is not that useful for plotting raw scientific data for 2 reasons.


Firstly, the lines in `plot()` do not plot the line of best fit, they simply connect the neighboring data points. This is often not very useful. For example, here is a noisy sine curve.

```python
import numpy as np   # Import numpy

xs = np.linspace(0, 2*np.pi, 100)          # Create a list of 100 x-values from 0 to 2pi
ys = np.sin(xs)                            # Calculate sin(x) for each x
ys = ys + 0.5*(np.random.rand(100) - 0.5)  # Add a random noise to sin(x), with a standard deviation of 0.5

plt.plot(xs, ys)
```

It's not very useful to connect all the points, because we don't see the underlying sine curve! This is why instead of using `plot()`, we [Fit models to our data](../5%20Curve%20Fitting)

Secondly, `plot()` does not allow us to plot error bars on the data points! All of real scientific measurements are uncertain, so we need to use a tool that can reflect our uncertainty on the graph.

[See the `errorbar` tutorial to check out how to plot errorbars!](2%20Errorbar%20Plot%20Example.ipynb)
